## Preprocessing

```python
# Importing the necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import tensorflow as tf

# Read the data from charity_data.csv
url = "https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv"
df = pd.read_csv(url)

# Display the first few rows of the dataframe
df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
df.drop(columns=["EIN", "NAME"], inplace=True)
```

```python
# Determine the number of unique values in each column.
unique_counts = df.nunique()
print(unique_counts)
```

```python
# Look at APPLICATION_TYPE value counts for binning
app_counts = df["APPLICATION_TYPE"].value_counts()
print(app_counts)
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
app_cutoff = 500
application_types_to_replace = list(app_counts[app_counts < app_cutoff].index)

# Replace in the dataframe
for app in application_types_to_replace:
    df['APPLICATION_TYPE'] = df['APPLICATION_TYPE'].replace(app, "Other")

# Check to make sure binning was successful
print(df['APPLICATION_TYPE'].value_counts())
```

```python
# Look at CLASSIFICATION value counts for binning
class_counts = df["CLASSIFICATION"].value_counts()
print(class_counts)
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
class_counts_gt1 = df["CLASSIFICATION"].value_counts()[df["CLASSIFICATION"].value_counts() > 1]
print(class_counts_gt1)
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
class_cutoff = 1000
classifications_to_replace = list(class_counts_gt1[class_counts_gt1 < class_cutoff].index)
# Replace in the dataframe
for cls in classifications_to_replace:
    df['CLASSIFICATION'] = df['CLASSIFICATION'].replace(cls, "Other")

# Check to make sure binning was successful
print(df['CLASSIFICATION'].value_counts())
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
df = pd.get_dummies(df)
```

```python
# Split our preprocessed data into our features and target arrays
X = df.drop(columns=["IS_SUCCESSFUL"])
y = df["IS_SUCCESSFUL"]

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Create a StandardScaler instance
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
## New Optimization Attempt 2

# Define the model - deeper neural net with more neurons and relu activation
nn_optimization_2 = tf.keras.models.Sequential()

# First hidden layer with more neurons and relu activation
nn_optimization_2.add(tf.keras.layers.Dense(units=256, activation="relu", input_dim=len(X_train_scaled[0])))

# Additional hidden layers with more neurons and relu activation
nn_optimization_2.add(tf.keras.layers.Dense(units=128, activation="relu"))
nn_optimization_2.add(tf.keras.layers.Dense(units=64, activation="relu"))

# Output layer with sigmoid activation
nn_optimization_2.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn_optimization_2.summary()
```

```python
# Compile the model with Adamax optimizer
nn_optimization_2.compile(loss="binary_crossentropy", optimizer="adamax", metrics=["accuracy"])
```

```python
# Train the model with a higher number of epochs
fit_model_optimization_2 = nn_optimization_2.fit(X_train_scaled, y_train, epochs=100, verbose=1)
```

```python
# Evaluate the model using the test data
model_loss_optimization_2, model_accuracy_optimization_2 = nn_optimization_2.evaluate(X_test_scaled, y_test, verbose=2)
print(f"Loss: {model_loss_optimization_2}, Accuracy: {model_accuracy_optimization_2}")
```

```python
# Export the model to an HDF5 file
nn_optimization_2.save("AlphabetSoupCharity_Optimization_2.h5")
```

It seems that the model's performance has improved slightly compared to the first attempt. The training accuracy is higher, and the validation accuracy is also higher, which suggests that the model might be generalizing better to the validation data. However, there is still room for improvement.
